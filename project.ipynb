```python
1.How many are there in each Team and the percentage splitting with respect to the total employees.
```

```python
#Calculate number of employees in each team
team_employees=ABC_company["Team"].value_counts()
print("Number of employees in each team:")
team_employees
```

```python
# Calculate the percentage splitting with respect to the total employees
percentage=(team_employees/len(ABC_company))*100
print("\nPercentage splitting with respect to total employees:")
percentage
```

```python
# import matplotlib.pyplot as plt

# Bar chart for the distribution of employees across teams
team_employees.plot(kind='bar', color='skyblue')
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Plotting the percentage splitting of employees across teams
plt.figure(figsize=(10, 10))
percentage.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Percentage Splitting of Employees Across Teams')
plt.ylabel('')  
plt.show()
```

```python
3. Identify the predominant age group among employees
```

```python
# Define age groups
bins = [0,20, 25, 30, 35, 40, 45, 50]
labels = ['0-20', '20-25','25-30', '30-35', '35-40', '40-45', '45-50']
ABC_company['age_group'] = pd.cut(ABC_company['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = ABC_company['age_group'].value_counts()

# Identify the predominant age group
predominant_age_group = age_group_counts.idxmax()

print(age_group_counts)
print(f"The predominant age group is {predominant_age_group}.")
```

```python
# Bar chart for the distribution of age groups
age_group_counts.plot(kind='bar', color='orange')
for i, v in enumerate(age_group_counts):
    plt.text(i, v + 0.05, str(v), ha='center')
plt.title('Distribution of Age Groups Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
 4.Discover which team and position have the highest salary expenditure.
```

```python
# Group by team and calculate the total salary expenditure for each team
team_salary_expenditure = ABC_company.groupby('Team')['Salary'].sum()
team_salary_expenditure
```

```python
# Group by position and calculate the total salary expenditure for each position
position_salary_expenditure = ABC_company.groupby('Position')['Salary'].sum()
position_salary_expenditure
```

```python
# Identify the team and position with the highest salary expenditure
highest_team_expenditure = team_salary_expenditure.idxmax()
highest_position_expenditure = position_salary_expenditure.idxmax()

print(f"Team with the highest salary expenditure: {highest_team_expenditure}")
print(f"Position with the highest salary expenditure: {highest_position_expenditure}")
```

```python
# Pie chart for team salary expenditure
plt.figure(figsize=(10, 10))
team_salary_expenditure.plot(kind='pie',autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Team Salary Expenditure')
plt.ylabel('')
plt.show()

# Pie chart for position salary expenditure
position_salary_expenditure.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Position Salary Expenditure')
plt.ylabel('')
plt.show()
```

```python
5. Investigate if there's any correlation between age and salary, and represent it visually. 
```

```python
import matplotlib.pyplot as plt

# Calculate the correlation between age and salary
correlation = ABC_company['Age'].corr(ABC_company['Salary'])

# Plotting the correlation
plt.scatter(ABC_company['Age'], ABC_company['Salary'])
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

print(f"The correlation between age and salary is {correlation:.2f}")
```

```python
correlation= ABC_company[['Age', 'Salary']].corr()
sns.heatmap(correlation, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation between Age and Salary')
plt.show()
```
